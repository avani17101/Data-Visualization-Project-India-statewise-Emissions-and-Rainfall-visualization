```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, output_file
from bokeh.models import Slider, HoverTool, CustomJS
```

```python
def generate_line(state):
    plot = figure(title='GreenHouse Emission VS Rainfall', x_axis_label='Year')
    df_state = df[df['state'] == state]
    #plot.line(df_state['year'], df_state['rainfall'], line_width=2, line_color='blue', legend='Rainfall')
    print(df_state['year'],df_state['nitrogen'])
    plot.line(df['year'], df_state['nitrogen'], line_width=2, line_color='orange', legend='CH4')
    hover = HoverTool()
    hover.tooltips = """
    <div style=padding=5px>Year:@year</div>
    <div style=padding=5px>Population:@rainfall</div>
    """
    plot.add_tools(hover)
    output_notebook()
    show(plot)
```

```python
generate_line('Rajasthan')
```
