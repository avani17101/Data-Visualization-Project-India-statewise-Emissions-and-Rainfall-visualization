```python
import geopandas as gpd
import pandas as pd
import json
from pandas_ods_reader import read_ods
```

```python
datafile = 'data/rainfall.csv'

dfr = pd.read_csv(datafile)
dfr.columns = ['state', 'year', 'rainfall',]
dfr.head()
```

```python
dfc = read_ods("data/CO2.ods", 1)
dfc.head()
```

```python
dfn = read_ods("data/N2O.ods", 1)
dfn.head()
```

```python
dfh = read_ods("data/CH4.ods", 1)
dfh.head()
```

```python
def preprocess(df,name):
    cols = ['State']
    for d in df.columns:
        x = df[d][0]
    #     print(x)
        if(x!='State'):
            y = x.split(" ")
            yr = y[4].split('-')
            cols.append(y[3].replace(",",""))
    df.columns = cols
    state_yr = []
    state = ''
    for i in range(2,len(df)):
        for col in cols:
            if(col =='State'):
                state = df[col][i]
            else:
                year = col
                val = df[col][i]
    #             print(state, year,val)
                state_yr.append([state, year,val])
    df = pd.DataFrame(state_yr)
    df.head()
    df.columns = ['state','year',name]
    df = delete_null(df)
    return df
```

```python
def delete_null(df):
    df = df.drop(df.index[0])
    df = df.dropna()
    df = df.reset_index(drop=True)
    return df
```
